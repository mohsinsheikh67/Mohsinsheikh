```python
import pandas as pd
```

```python
data = {'Job Title': 'python developer', 'Company': ['Hidragen technologies', 'The Scalers', 'Intelligent outsourcing'], 'Location': ['Jaipur', 'Bangalore', 'Mohali'], 'Salary': ['60000', '50000', '40000']}
df = pd.DataFrame(data)
```

```python
df.to_excel('python_developer_jobs.xlsx', index=False)
```

```python
import numpy as np
import matplotlib.pyplot as plt

```

```python
df = pd.read_excel('python_developer_jobs.xlsx')
average_salary = df.groupby('Location')['Salary'].mean()
```

```python
plt.bar(average_salary.index, average_salary.values)
plt.xlabel('City')
plt.ylabel('Average Salary')
plt.title('Average Salary for Python Developers Across Cities')
plt.xticks(rotation=45)
plt.show()
```
